```python
# 1. Create 1-D and 2-D arrays using Numpy. Also convert 1-D array to appropriate 2-D using reshape() and add two arrays. Also find the shape of a 2-D array using shape().
arr_1d = np.array([1,2,3,4,5,6])
print(arr_1d)
arr_2d = np.array([[7,8,9],[10,11,12]])
print(arr_2d)
```

```python
# Convert 1-D array to appropriate 2-D using reshape() and add two arrays.
new_2d = arr_1d.reshape(2,3)
print(new_2d)

added_arr = new_2d + arr_2d
print(added_arr)

print(added_arr.shape)
```

```python
# 2. Use an appropriate indexing method to access the last data of the 2-D array.
ar = np.array([[1,2,3],[4,5,6]])
print(ar[-1,-1])  #Negative Indexing  (Last data access)
print(ar[1,2])
```

```python
# 3. For the above 1-D and 2-D array created, slice the arrays to get specific data you want. You can also use negative slicing.
print(ar[0,0:1])
print(ar[0:1,0:1])
```

```python
# 4. Search the desired data using where() and searchsorted()
arrr = np.array([4,5,6,7,8,9,10])
x = np.where(arrr==5)
print("at",x)

ans = np.searchsorted(arrr, 5)
print(ans)
```

```python
# 5. Calculate mean, median, mode, standard deviation and variance for a 1-D data. Also find the MSE and RMSE.
print("Mean =", np.mean(arrr))
print("Median =", np.median(arrr))
print("Standard Deviation =", np.std(arrr))
print("Variance =", np.var(arrr))

actualmarks = np.array([1,2,3,4,5])
predictedmarks = np.array([5,4,3,2,1])
mse = np.mean((actualmarks - predictedmarks)**2)
rmse = np.sqrt(mse)
print("MSE =",mse)
print("RMSE =",rmse)
```

```python
# 9. Create a line plot, scatter plot, bar graph and pie chart with labels using matplotlib for a given series of data
# Subjects Marks
# Probability and Statistics 60
# Data Science and Machine Learning 45
# Artificial Intelligence 59
# Computer Graphics 72
# Engineering Economics 80
# Research Fundamentals 91

import matplotlib.pyplot as plt
subjects = ['Probability and Statistics', 'Data Science and Machine Learning', 'Artificial Intelligence', 'Computer Graphics', 'Engineering Economics', 'Research Fundamentals']
marks = [60, 45, 59, 72, 80, 91]
# Line Plot
plt.plot(subjects, marks, marker='o')
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot
plt.scatter(subjects, marks, color='red')
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
#Bar Graph
plt.bar(subjects, marks, color='blue') 
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart
plt.pie(marks, labels=subjects)
plt.title('Marks Distribution in Subjects')
plt.show()
```
