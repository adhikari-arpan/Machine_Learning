```python
import pandas as pd
import numpy as np
```

```python
data = {
    'area' : [50, 70, 75, 85, 90, 101],
    'price' : [200, 280, 325, 390, 378, 385]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['area']]
y = df['price']

model = LinearRegression()
model.fit(X,y)
```

```python
a = model.intercept_
b = model.coef_

print(f"Y = {a:.2f}+{b[0]:.2f}X")
```

```python
# Find MSE and R2 Score.
y_pred = model.predict(X)
df['predicted'] = y_pred
df
```

```python
MSE = np.mean((df['price'] - df['predicted'])**2)
print(f"MSE = {MSE:.2f}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
MSE = mean_squared_error(df['price'], df['predicted'])
print(f"MSE = {MSE:.2f}")
r2 = r2_score(df['price'], df['predicted'])
print(f"R2 Score = {r2:.2f}")
# R2 Score gives goodness of fit
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['area'], df['price'])
plt.plot(df['area'], df['predicted'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression')
plt.show()
```
