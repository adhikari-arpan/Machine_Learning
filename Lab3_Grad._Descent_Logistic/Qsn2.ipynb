```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Study_hrs' : [15,29,28,31,35],
    'Result' : ['Fail', 'Fail', 'Pass', 'Pass', 'Pass']
}

df = pd.DataFrame(data)
df
```

```python
X = df['Study_hrs']
y = df['Result']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train.values.reshape(-1,1), y_train)
# reshape(-1,1) means 1 column exactly but rows = total number of elements present

y_pred = model.predict(X_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
y_pred_prob = model.predict_proba(X.values.reshape(-1,1))
plt.scatter(X, y)
plt.plot(X, y_pred_prob[:,1])
plt.show()
```
