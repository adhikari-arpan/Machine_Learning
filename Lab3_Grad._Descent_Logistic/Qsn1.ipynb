```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = {
    'X': [1,2,3,4,5],
    'Y': [2,5,7,8,9]
}

df = pd.DataFrame(data)
df
```

```python
X = df['X']
y = df['Y']

a = 0
b = 0
alpha = 0.01
n = len(X)
```

```python
for i in range(100):
  y_pred = a + b * X
  error = y - y_pred
  mse = mean_squared_error(y, y_pred)
  print("MSE :", mse)
  a_grad = -2 * np.sum(error) / n
  b_grad = -2 * np.sum(error * X) / n
  a = a - alpha * a_grad
  b = b - alpha * b_grad
```

```python
print("Optimized Equation is")
print(f"y = {a:.2f} + {b:.2f}X")
print(f"\nMSE = {mse:.3f}")
print(f"R2 Score = {r2_score(y,y_pred):.3f}")
```
